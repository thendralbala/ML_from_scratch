#Data Preprocessing Techniques



Data preprocessing is the process of transforming raw data into a format that is more suitable for machine learning algorithms. This is a crucial step in the machine learning pipeline, as the quality of the data can significantly impact the performance of the model.

Here are some common data preprocessing techniques:

1. Data Cleaning:

Handling missing values: Missing values can be filled using techniques like imputation (e.g., replacing with mean, median, or mode) or removing rows/columns with missing values.
Handling noisy data: Noisy data (e.g., outliers or errors) can be smoothed or removed using techniques like binning, regression, or outlier detection.
Data deduplication: Removing duplicate data points from the dataset.
2. Data Transformation:

Scaling: Transforming features to a specific range, such as 0 to 1 or -1 to 1, using techniques like standardization (z-score normalization) or min-max scaling.
Normalization: Adjusting the distribution of data to a normal distribution.
Encoding categorical variables: Converting categorical variables (e.g., colors, genders) into numerical representations using techniques like one-hot encoding or label encoding.
3. Data Reduction:

Feature selection: Identifying the most relevant features for the machine learning task and discarding irrelevant or redundant features. This helps reduce dimensionality and improve model performance.
Dimensionality reduction: Using techniques like Principal Component Analysis (PCA) or Linear Discriminant Analysis (LDA) to reduce the number of features while preserving important information.
Reasoning:

These preprocessing steps are essential for several reasons:

Improved model accuracy: By cleaning and transforming the data, machine learning algorithms can learn patterns more effectively, leading to better predictions.
Reduced overfitting: Preprocessing can help prevent overfitting, where the model learns the training data too well and performs poorly on unseen data.
Faster training: Preprocessed data can often be processed more efficiently by machine learning algorithms, reducing training time.

**Why is NumPy important in machine learning and data science?**

 NumPy allows you to perform operations on entire arrays at once, rather than looping through individual elements. This "vectorization" significantly improves performance, especially for numerical computations.

```python
import numpy as np
from sklearn import preprocessing
```

```python
#Creating sample data
data = np.array([[3,1.5,2,-5.4],
                 [0,4,-0.3,2.1],
                 [1,3.3,-1.9,-4.3]])
```

**Mean Removal**



*   The *scale* function is used for standardization, a common data preprocessing technique. It centers the data by subtracting the mean of each feature and then scales it by dividing by the standard deviation of each feature. This transforms the data to have zero mean and unit variance.
*   `data_standardized.mean(axis = 0)`: This calculates the mean of the standardized data along the columns (axis 0). In other words, it calculates the mean for each feature.
*   `print(f"Std deviation = { ... }")`: This line prints the calculated standard deviation, which should be very close to 1 for each feature after standardization.


Standardization is often used in machine learning to ensure that features with different scales do not disproportionately influence the model's learning process. It helps algorithms converge faster and can improve the performance of many models




```python
data_standardized = preprocessing.scale(data)
print(f"Mean = {data_standardized.mean(axis = 0)}")
print(f"Std deviation = {data_standardized.std(axis = 0)}")
```

**Scaling**

In data preprocessing, scaling refers to techniques that transform the features of a dataset to a specific range. This is often done to ensure that features with different scales (e.g., one feature ranging from 0 to 100, while another ranges from 0 to 1) do not disproportionately influence the learning process of machine learning algorithms.

There are several common scaling techniques, including:

Standardization: Transforming data to have a mean of 0 and a standard deviation of 1.

Min-Max Scaling: Transforming data to a specific range, typically between 0 and 1.



*   `preprocessing.MinMaxScaler(feature_range=(0,1))`: This creates an instance of the MinMaxScaler class from the sklearn.preprocessing module. MinMaxScaler is used for min-max scaling.

    `feature_range=(0,1)`: This argument specifies the desired range for the scaled data, which in this case is between 0 and 1 (inclusive).
*   `fit`: This calculates the minimum and maximum values of each feature in the input data (data) and stores them within the data_scaler object.

    `transform`: This applies the min-max scaling transformation using the calculated minimum and maximum values to the input data (data).



```python
data_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
data_scaled = data_scaler.fit_transform(data)
print(f"Min max scaled data = \n {data_scaled}")
```

**Normalization**

In data preprocessing, normalization refers to techniques that adjust the values of a dataset to a common scale, typically without distorting differences in the ranges of values or losing information. It's often used to ensure that features with different units or scales contribute equally to the analysis.

One common type of normalization is L1 normalization, which scales each sample (row) of the data so that the sum of the absolute values of its elements is 1. This is also known as Manhattan normalization.

`preprocessing.normalize(data, norm='l1')`: This line uses the normalize function from the sklearn.preprocessing module to apply L1 normalization to the input data (data).

L1 normalization is often used in text classification and clustering where the sparsity of the data is important. It's also useful when you want to ensure that features with different units or scales contribute equally to the analysis, without distorting the differences in the ranges of values.

```python
data_normalized = preprocessing.normalize(data, norm='l1')
print(f"L1 Normalized Data = \n{data_normalized}")
```

**Binarization**

Data binarization is a preprocessing technique where you transform numerical data into binary values (0 or 1). This is done by applying a threshold: if a data point is above the threshold, it's assigned a value of 1; otherwise, it's assigned a value of 0.

```python
data_binarized = preprocessing.Binarizer(threshold=1.4).transform(data)
print(f"Binarized Data = \n {data_binarized}")
```

**One Hot Encoding**

One-hot encoding is a technique used to convert categorical data into a numerical format that machine learning algorithms can understand. It works by creating new binary features (columns) for each unique category in the original categorical feature. These new features are assigned a value of 1 if the original data point belongs to that category and 0 otherwise.

In this specific scenario, the encoder will create one-hot encoded features for each of the four input features. The length of the encoded vector will depend on the number of unique values observed in each feature during the fitting process.

**Example:**

Let's assume the following unique values were observed in each feature during fitting:

Feature 1: [0, 1, 2]

Feature 2: [2, 3]

Feature 3: [1, 3, 4, 5]

Feature 4: [3, 12]

Then, the one-hot encoded vector for the input data point [2, 3, 5, 3] would likely look something like this:

[0, 0, 1, 0, 1, 0, 0, 0, 1, 0, 1, 0]

The first three elements represent Feature 1 (0, 1, 2), with the third element (1) indicating the presence of value 2.

The next two elements represent Feature 2 (2, 3), with the second element (1) indicating the presence of value 3.

The following four elements represent Feature 3 (1, 3, 4, 5), with the fourth element (1) indicating the presence of value 5.

The last two elements represent Feature 4 (3, 12), with the first element (1) indicating the presence of value 3.

**Reasoning:**

One-hot encoding is crucial for categorical data as it allows machine learning algorithms to handle categorical features without imposing any ordinal relationship between the categories. This ensures that the algorithm doesn't misinterpret the data and can learn patterns effectively.

```python
encoder = preprocessing.OneHotEncoder()
encoder.fit([[0, 2,1, 12],
             [1, 3, 5, 3],
             [2, 3, 3, 12],
             [1, 2, 4, 3]])
encoded_vector = encoder.transform([[2, 3, 5, 3]]).toarray()
print(f"Encoded vector = \n {encoded_vector}")
```

#Label Encoding

Label encoding is a simple technique used to convert categorical data into numerical format. It works by assigning a unique integer to each category in a categorical feature. These integers are typically assigned in alphabetical order or based on the frequency of each category.

```python
label_encoder = preprocessing.LabelEncoder()
input_classes = ['audi', 'ford', 'audi', 'toyota', 'ford', 'bmw']
label_encoder.fit(input_classes)

print(f"Class mapping: ")
for i, item in enumerate(label_encoder.classes_):
    print(f"{item} --> {i}")
```

```python
labels = ['toyota', 'ford', 'audi']
encoded_labels = label_encoder.transform(labels)
print(f"Labels: {labels}")
print(f"Encoded Labels: {encoded_labels}")
```

```python
encoded_labels = [2,1,0,3,1]
decoded_labels = label_encoder.inverse_transform(encoded_labels)
print(f"Encoded Labels: {encoded_labels}")
print(f"Decoded Labels: {decoded_labels}")
```

#Building a Linear Regressor

Linear Regression is a supervised machine learning algorithm used for predicting a continuous target variable based on one or more predictor variables. It assumes a linear relationship between the target and predictor variables.

```python
import sys
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
#file downloaded from: https://www.kaggle.com/datasets/gurdit559/canada-per-capita-income-single-variable-data-set
filename = "canada_per_capita_income.csv"
X = []
Y = []

with open(filename,'r') as f:
  next(f, None)
  for line in f.readlines():
    x,y = line.split(',')
    X.append(float(x))
    Y.append(float(y))

X = np.array(X)
Y = np.array(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.20, random_state=0)

X_train = X_train.reshape(-1,1)
Y_train = Y_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
```

```python
linear_regressor = linear_model.LinearRegression()

#Train the model using the training sets
linear_regressor.fit(X_train,Y_train)
```

```python
y_train_pred = linear_regressor.predict(X_train)
plt.figure()
plt.scatter(X_train,Y_train, color = 'green')
plt.plot(X_train, y_train_pred, color = 'black', linewidth = 4)
plt.title('Training Data')
plt.show()
```

```python
y_test_pred = linear_regressor.predict(X_test)

plt.scatter(X_test, Y_test, color = 'green')
plt.plot(X_test, y_test_pred, color = 'black', linewidth = 4)
plt.title("Test data")
plt.show()
```

#Computing Regression Accuracy

1. Mean Absolute Error (MAE):

Definition: The average of the absolute differences between predicted and actual values.
Interpretation: Represents the average error magnitude, without considering direction. Lower values indicate better accuracy.

Formula: MAE = (1/n) * Σ|yᵢ - ŷᵢ| (where n is the number of data points, yᵢ is the actual value, and ŷᵢ is the predicted value).
2. Mean Squared Error (MSE):

Definition: The average of the squared differences between predicted and actual values.
Interpretation: Penalizes larger errors more heavily than MAE. Sensitive to outliers. Lower values indicate better accuracy.

Formula: MSE = (1/n) * Σ(yᵢ - ŷᵢ)²
3. Median Absolute Error (MedAE):

Definition: The median of the absolute differences between predicted and actual values.

Interpretation: Robust to outliers, less affected by extreme errors compared to MAE. Lower values indicate better accuracy.
4. Explained Variance Score:

Definition: Measures the proportion of variance in the target variable explained by the model.

Interpretation: Higher values (closer to 1) indicate a better fit, meaning the model explains more of the variation in the data.

Formula: 1 - (Var(y - ŷ) / Var(y)) (where Var represents variance).
5. R-squared (R²):

Definition: Similar to explained variance, but can be negative for poor models. Represents the goodness of fit.

Interpretation: Higher values (closer to 1) indicate a better fit. Values closer to 0 suggest the model is no better than predicting the mean. Negative values indicate a very poor fit.

Formula: 1 - (Σ(yᵢ - ŷᵢ)² / Σ(yᵢ - ȳ)²) (where ȳ is the mean of actual values).

```python
import sklearn.metrics as sm
```

```python
print("Mean Absolute Error = ", round(sm.mean_absolute_error(Y_test, y_test_pred),2))
print("Mean Squared Error = ", round(sm.mean_squared_error(Y_test, y_test_pred),2))
print("Median Absolute Error = ", round(sm.median_absolute_error(Y_test, y_test_pred), 2))
print("Explained Variance Score = ", round(sm.explained_variance_score(Y_test, y_test_pred), 2))
print("R2 Score = ", round(sm.r2_score(Y_test, y_test_pred), 2))
```

#Building a Ridge Regressor

Ridge Regression is used to address multicollinearity in linear regression. Multicollinearity occurs when predictor variables in your dataset are highly correlated. This can make the model unstable and lead to unreliable coefficient estimates.

Ridge Regression mitigates this issue by adding a penalty term (L2 regularization) to the loss function. This penalty shrinks the coefficients towards zero, reducing their variance and improving model stability.


This leads to:

More robust models: Less sensitive to fluctuations in the data.

Better generalization: Improved performance on unseen data by reducing overfitting.

Handling multicollinearity: Allows using correlated features without drastic impact on model accuracy.




Ridge Regression doesn't directly identify or remove outliers. Instead, it makes the model less sensitive to them by reducing their influence on the model's fitting process. This indirect effect can help improve the model's overall performance, especially in the presence of outliers.

`alpha = 0.01:`

alpha is the regularization strength parameter. It controls the amount of shrinkage applied to the coefficients.

A higher alpha value leads to stronger regularization, shrinking the coefficients more aggressively towards zero.

A lower alpha value results in weaker regularization, allowing the coefficients to have larger magnitudes.
In this case, alpha = 0.01 indicates a relatively weak regularization.

`fit_intercept = True:`

This parameter determines whether to calculate the intercept (bias term) for the model.

When set to True (default), the model will fit an intercept, allowing the regression line to shift vertically.

Setting it to False forces the regression line to pass through the origin (0, 0).

`max_iter = 10000:`

This parameter specifies the maximum number of iterations for the solver to converge to a solution.
The solver is an iterative algorithm used to find the optimal coefficients.

Setting a higher max_iter value allows the solver more attempts to converge, potentially leading to a more accurate solution.

In this case, max_iter = 10000 provides a generous limit for the solver to find the best coefficients.

```python
ridge_regressor = linear_model.Ridge(alpha = 0.001, fit_intercept = True, max_iter = 10000)
```

```python
ridge_regressor.fit(X_train, Y_train)
y_test_pred_ridge = ridge_regressor.predict(Y_test)
```

```python
print("Mean Absolute Error = ", round(sm.mean_absolute_error(Y_test, y_test_pred_ridge), 2))
print("Mean Squared Error = ", round(sm.mean_squared_error(Y_test, y_test_pred_ridge), 2))
```

#Estimating Housing Prices

 **Decision Tree Regressors**

A Decision Tree Regressor is a supervised machine learning algorithm used for predicting a continuous target variable. It works by recursively partitioning the data into subsets based on the values of input features, creating a tree-like structure. Each internal node of the tree represents a decision based on a feature, and each leaf node represents a predicted value for the target variable.

***How it Works***


*   Tree Construction: The algorithm starts with the entire dataset and selects the best feature to split the data based on a criterion like minimizing variance or mean squared error. This process is repeated recursively for each subset until a stopping condition is met (e.g., maximum depth, minimum samples per leaf).
*   Prediction: To predict the value for a new data point, the algorithm traverses the tree from the root node to a leaf node based on the values of the input features. The predicted value is the average or median of the target variable values in the leaf node.

***Advantages***

*   Easy to understand and interpret: The tree structure is visually intuitive and can be easily explained.
*   Handles both numerical and categorical data: Can be used with a mix of data types.
*   Non-parametric: Does not make assumptions about the underlying data distribution.
*   Feature selection: Automatically identifies important features for prediction.

***Disadvantages***

*   Prone to overfitting: Can create complex trees that memorize the training data, leading to poor generalization on unseen data.
*   Sensitive to small changes in data: Minor variations in the data can result in different tree structures.
*   Can be unstable: Predictions can vary significantly with small changes in the data.









**AdaBoost Regressor with Decision Tree**

AdaBoost (Adaptive Boosting) is an ensemble learning method that combines multiple weak learners (often Decision Trees) to create a strong learner. In the context of regression, AdaBoost Regressor aims to improve the accuracy of predictions by sequentially training weak learners, each focusing on the errors made by the previous ones.

***How it Works***
*   Initialization: Initially, all data points are assigned equal weights.
*   Weak Learner Training: A weak learner (e.g., a Decision Tree with limited depth) is trained on the data.
*   Weight Adjustment: Data points that were mispredicted by the weak learner are given higher weights, while correctly predicted points receive lower weights.
*   Iteration: Steps 2 and 3 are repeated for a specified number of iterations, creating a sequence of weak learners.
*   Ensemble Prediction: The final prediction is a weighted average of the predictions from all weak learners, where the weights are determined by their individual performance.

***Using Decision Trees as Weak Learners***

Decision Trees are commonly used as weak learners in AdaBoost because they are relatively simple to train and can capture non-linear relationships in the data. By combining multiple Decision Trees with varying depths and structures, AdaBoost can create a more robust and accurate regressor.

***Advantages***



*   Improved accuracy: AdaBoost often achieves higher accuracy compared to individual weak learners.
*   Robustness to overfitting: By combining multiple weak learners, it reduces the risk of overfitting.
*   List itemHandles complex relationships: Can model non-linear patterns in the data.

***Disadvantages***

*   Sensitive to noisy data: Outliers can significantly impact the performance.
*   Computationally intensive: Training multiple weak learners can take longer.















```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn import datasets
from sklearn.metrics import mean_squared_error, explained_variance_score
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Download housing dataset
from sklearn.datasets import fetch_california_housing
housing_data = fetch_california_housing()

# Input dataset can be accessed using housing_data.data and corresponding price using housing_data.target

X, y = shuffle(housing_data.data, housing_data.target, random_state = 7) #The random state parameter controls
                                                                         #how the data is shuffled so that we
                                                                         #can have reproducible results

X_train,X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
dt_regressor = DecisionTreeRegressor(max_depth = 10)
dt_regressor.fit(X_train, y_train)
```

```python
ab_regressor = AdaBoostRegressor(DecisionTreeRegressor(max_depth = 10), n_estimators=1000, random_state = 7)
ab_regressor.fit(X_train, y_train)
```

```python
y_pred_dt = dt_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred_dt)
evs = explained_variance_score(y_test, y_pred_dt)
print("Decision Tree Performance:")
print(f"Mean Squared Error = {round(mse,2)}")
print(f"Explained Variance Score = {round(evs,2)}")
```

```python
y_pred_ab = ab_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred_ab)
evs = explained_variance_score(y_test, y_pred_ab)
print("AdaBoost Performance:")
print(f"Mean Squared Error = {round(mse,2)}")
print(f"Explained Variance Score = {round(evs,2)}")
```

#Computing the Relative Importance of Features

The regressor object has a callabe `feature_importances_ ` method that gives us the relative importance of each feature

```python
def plot_feature_importances(feature_importances, title, feature_names):
  #Normalize the importance values
  feature_importances = 100.0 * (feature_importances / max(feature_importances))
  index_sorted = np.flipud(np.argsort(feature_importances))

  #center the location of labels on the X- axis  (for display purposes)
  pos = np.arange(index_sorted.shape[0]) + 0.5

  #Plot bar graph
  plt.figure(figsize=(12, 4))
  plt.bar(pos, feature_importances[index_sorted], align = 'center')
  plt.xticks(pos, np.array(feature_names)[index_sorted])
  plt.ylabel('Relative Importance')
  plt.title(title)
  plt.show()
```

```python
plot_feature_importances(dt_regressor.feature_importances_, 'Decision Tree Regressor', housing_data.feature_names)
plot_feature_importances(ab_regressor.feature_importances_, 'AdaBoost Regressor', housing_data.feature_names)
```

```python
#Dataset downloaded from https://archive.ics.uci.edu/dataset/275/bike+sharing+dataset

import csv
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.metrics import mean_squared_error, explained_variance_score
import matplotlib.pyplot as plt

```

```python
def load_dataset(filename):
  file_reader = csv.reader(open(filename,'r'), delimiter = ',')
  X,y = [], []
  for row in file_reader:
    X.append(row[2:13])
    y.append(row[-1])

    #Extract Feature Names
    feature_names = np.array(X[0])

    #Remove the first row because they are feature names
  return np.array(X[1:]).astype(np.float32),np.array(y[1:]).astype(np.float32),feature_names
```

```python
X,y,feature_names = load_dataset("day.csv")
X, y = shuffle(X, y, random_state = 7)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.1, random_state = 0)
```

```python
rf_regressor = RandomForestRegressor(n_estimators = 1000, max_depth = 10, min_samples_split = 2)
rf_regressor.fit(X_train, y_train)
```

```python
y_pred = rf_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
evs = explained_variance_score(y_test, y_pred)
print("Random Forest Regressor Performance:")
print("Mean Squared Error = ", round(mse,2))
print("Explained Variance Score = ", round(evs, 2))
```
